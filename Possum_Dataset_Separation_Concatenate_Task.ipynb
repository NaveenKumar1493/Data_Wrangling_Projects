# Seprate Rows by Percentage and Column by Numbers 

```python
y1=df.iloc[0:10,0:2]           #10%   
y2=df.iloc[10:16,0:2]          #5%
y3=df.iloc[0:16,2:6]           #15%
y4=df.iloc[16:36,0:6]          #20%    
y5=df.iloc[0:36,6:9]           #35%
y6=df.iloc[0:21,9]             #20%
y7=df.iloc[21:36,9]            #15%
y8=df.iloc[0:31,10:13]         #30%
y9=df.iloc[31:42,10:13]        #10%
y10=df.iloc[36:42,0:10]        #5%

# at this stage data will divided in to 40% from (0 t0 12) column

                          
y11=df.iloc[42:94,0:3]         #50%
y12=df.iloc[94:99,0]           #5%
y13=df.iloc[94:99,1:3]         #5%
y14=df.iloc[42:78,3:6]         #35%
y15=df.iloc[78:99,3]           #20%
y16=df.iloc[78:88,4:6]         #10%
y17=df.iloc[88:99,4:6]         #10%
# at this stage data will divided in to 95% from (0 t0 5) column

y18=df.iloc[42:67,6:13]       #25%
y19=df.iloc[67:99,6]          #30%
# at this stage data will divided in to 95% from (0 t0 6) column



y20=df.iloc[67:99,7:13]       #30%
# at this stage data will divided in to 95% from (0 t0 12) column

y21=df.iloc[0:99,13]         #95%
# at this stage data will divided in to 95% woith (0 t0 13) column                  

y22=df.iloc[99:,0:14]         #5%
# at this stage data will divided in to 100% woith (0 t0 13) column


```

# Concatenate the Whole Dataset

```python
# Merging the data

#Phase1

merge = pd.concat([y1,y2],axis=0)
merge1 = pd.concat([merge,y3],axis=1)
merge2 = pd.concat([merge1,y4],axis=0)
merge3 = pd.concat([merge2,y5],axis=1)
joint = pd.concat([y6,y7],axis=0)
merge4 = pd.concat([merge3,joint],axis=1)
merge5=pd.concat([merge4,y10],axis=0)
joint_1 = pd.concat([y8,y9],axis=0)
merge6 = pd.concat([merge5,joint_1],axis=1)


#Phase2

joint_2 = pd.concat([y12,y13],axis=1)
merge7 = pd.concat([y11,joint_2],axis=0)
joint_3 = pd.concat([y16,y17],axis=0)
joint_4 = pd.concat([y15,joint_3],axis=1)
joint_5 = pd.concat([y14,joint_4],axis=0)
merge8 = pd.concat([merge7,joint_5],axis=1)
joint_6 = pd.concat([y19,y20],axis=1)
joint_7 = pd.concat([y18,joint_6],axis=0)
merge9 = pd.concat([merge8,joint_7],axis=1)
merge10 = pd.concat([merge6,merge9],axis=0)
merge11 = pd.concat([merge10,y21],axis=1)
merge12 = pd.concat([merge11,y22],axis=0)
merge12
```
